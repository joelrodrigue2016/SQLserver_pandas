


Harold Frederick Shipman, Jr. (14 January 1946 – 13 January 2004), known to acquaintances as Fred Shipman, was an English general practitioner and serial killer. He is considered to be one of the most prolific serial killers in modern history, with an estimated 250 victims. On 31 January 2000, Shipman was found guilty of murdering fifteen patients under his care. He was sentenced to life imprisonment with a whole life order. Shipman died by suicide by hanging himself in his cell at HM Prison Wakefield, West Yorkshire, on 13 January 2004, aged 57.
More details here: https://en.wikipedia.org/wiki/Harold_Shipman

This part of the dataset describes the gender of the victims that Dr. Shipman killed.

```python
query = """
SELECt age, gender2,count(*) as amount
  FROM [practical_sql].[dbo].[shipman_confirmed_victims]
  group by age,gender2
  order by count(*) desc


  
 

"""
df = pd.read_sql_query(query, conn)


df
```

```python
import plotly.express as px

gender= df['gender2']
age=df['age']



fig = px.pie(values=age, names=gender)
fig.update_layout(
    title=dict(text="Shipman's prefered victims were Women"), font=dict(size=15)
)
fig.update_layout(paper_bgcolor="lightblue")
fig.show()
```

Most of Shipman's victims were killed in the 1990s.

```python
query = """
SELECT yearofDeath,count(yearofDeath) as number
  FROM [practical_sql].[dbo].[shipman_confirmed_victims]
  group by yearofDeath



  
 

"""
df = pd.read_sql_query(query, conn)
df
```

```python
import numpy
import plotly.express as px



# Random Data
names = df['yearofDeath']
x =   df['number']

fig = px.pie(values=x, names=names)
fig.update_layout(
    title=dict(text="Years when Shipman killed most of his victims."),font=dict(size=15)
)
fig.update_layout(paper_bgcolor="lightblue")
fig.show()
```

Shipman preferred place to kill his victims was their own home.

```python
query = """
SELECT placeofdeath, count(placeofdeath) as [amount of deaths]
  FROM [practical_sql].[dbo].[shipman_confirmed_victims]
  group by PlaceofDeath
  order by count(placeofdeath) desc


  
 

"""
df = pd.read_sql_query(query, conn)
df
```

```python
import numpy
import plotly.express as px


# Random Data
names = df['placeofdeath']
x = df['amount of deaths']

fig = px.pie(values=x, names=names)
fig.update_layout(
    title=dict(text="Shipman killed most of his victims in their own home."), font=dict(size=15)
)
fig.update_layout(paper_bgcolor="lightblue")
fig.show()
```

Most of the Shipman's victims were between the ages of 60-85.

```python
query = """
SELECt age, gender2,count(*) as amount
  FROM [practical_sql].[dbo].[shipman_confirmed_victims]
  group by age,gender2
  order by count(*) desc



  
 

"""
df = pd.read_sql_query(query, conn)
df
```

```python
import plotly.express as px

age, gender = df["age"],df["gender2"]


 
fig = px.box(y = age, x=gender, points="all")
fig.update_layout(
    title=dict(text="Shipman's preferred victims were old old people"),font=dict(size=15)
)
fig.update_layout(paper_bgcolor="lightpink",plot_bgcolor = "rgba(0,0,0,0)")
fig.update_layout(

    xaxis_title="Months",
    yaxis_title="Years ",
    legend_title="Legend Title"
)



fig.show()
```

```python
query = """
SELECT YEAR([DateofDeath]) AS Year,
	   FORMAT([DateofDeath], 'MMMM') AS Month,
	   day([DateofDeath]) AS Day,
	   *         

 from [dbo].[shipman_confirmed_victims]
 


  
 

"""
df = pd.read_sql_query(query, conn)
df
```

Shipman killed most of his victims diring fall time, especially December.

```python

month=df["Month"]
year = df["Year"]
fig = px.box(x=month, y=year)
fig.update_layout(title=dict(text="Shipman's killings per months in a spam or more than 3 decades"), font=dict(size=15))
fig.update_layout(paper_bgcolor="lightpink", plot_bgcolor="rgba(0,0,0,0)")


fig.update_layout(
   
    xaxis_title="Months",
    yaxis_title="Years ",
    legend_title="Legend Title"
    )






fig.show()
```
